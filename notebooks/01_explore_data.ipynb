````python
import sys
from pathlib import Path

# Make project root importable from a notebooks/ subfolder
PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import pandas as pd
import matplotlib.pyplot as plt

from config.settings import MONTHLY_INSURANCE_PAID_FILE, MONTHLY_DENIALS_FILE
````

````python
paid_df = pd.read_parquet(MONTHLY_INSURANCE_PAID_FILE)
den_df = pd.read_parquet(MONTHLY_DENIALS_FILE)
paid_df.head()
````

````python
# Monthly insurance paid by insurance_type
fig, ax = plt.subplots(figsize=(10, 5))
for ins, g in paid_df.groupby('insurance_type'):
    monthly = g.groupby('month')['total_insurance_paid'].sum()
    monthly.plot(ax=ax, label=ins, marker='o')
ax.set_title('Monthly Insurance Paid by Insurance Type')
ax.set_ylabel('Insurance Paid')
ax.legend()
plt.tight_layout()
plt.show()
````

````python
# Monthly denials by financial_class
fig, ax = plt.subplots(figsize=(10, 5))
for fc, g in den_df.groupby('financial_class'):
    monthly = g.groupby('month')['total_denials'].sum()
    monthly.plot(ax=ax, label=fc, marker='o')
ax.set_title('Monthly Denials by Financial Class')
ax.set_ylabel('Denial Amount')
ax.legend()
plt.tight_layout()
plt.show()
````

````python
# Seasonality (month-of-year average across years)
paid_df['month_num'] = pd.to_datetime(paid_df['month']).dt.month
seasonal = paid_df.groupby('month_num')['total_insurance_paid'].mean()
seasonal.plot(kind='bar', figsize=(10, 4), title='Average Insurance Paid by Month-of-Year')
plt.show()
````
